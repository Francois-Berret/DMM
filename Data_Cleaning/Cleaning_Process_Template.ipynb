```python
#Creation of a "skills" column from the skills mentioned in each job description

import numpy as np

skills1 = ["Python", "SQL", "AWS", "Spark", "Azure", " R ", "Tableau", "Java", "Excel", "Scala", "Hadoop", "Power BI", "Snowflake", "Kafka", "NoSQL", "DataBricks", "Redshift",
          "Git", "Oracle", "Airflow", "GCP", "Docker", "SQL Server", "SAS", "Kubernetes", "BI ", "Data Viz", "data visualisation", "Machine learning", "Machine Learning",
           "Deep learning", "Deep Learning" "Dashboard", "Postgresql", "API", "Big data", "Big Data", "ETL", "Agile", "CRM", "Devops", "DevOps", "d√©ploiement", "algorithmes"]

skills2 = [skill.lower() for skill in skills1]
skills = skills1 + skills2

dfW["global_job_desc"] = dfW["job_description"] + dfW["profile"] #Creation of a column "global_job_desc" grouping "job_description" and "profile"
column = "global_job_desc"

def found_skills(description):
    skills_found = []
    for skill in skills:
        if skill in description:
            skills_found.append(skill)
    return skills_found if skills_found else np.nan

dfW['global_job_desc'] = dfW['global_job_desc'].apply(lambda x: str(x))
dfW['profile'] = dfW['profile'].apply(lambda x: str(x))
dfW['skills'] = dfW[column].apply(found_skills)
dfW['skills'] = dfW['skills'].str.join(', ')

dfW
```

```python
#Import of a dataset containing the names of all the cities and regions of France

fr_cities_df = pd.read_excel('/content/drive/MyDrive/Jedha_Fullstack/DataMatch_Project/Data_Cleaning/Liste-Excel-des-villes-de-France.xlsx')
```

```python
fr_cities_df.head()
```

```python
fr_cities_list = fr_cities_df.iloc[2:,0]
fr_regions_list = fr_cities_df.iloc[2:,3]
fr_dept_list = fr_cities_df.iloc[2:,4]
fr_list = pd.concat([fr_cities_list, fr_regions_list, fr_dept_list])
fr_list = fr_list.append(pd.Series(["La D√©fense", "France", "Sophia Antipolis", "Monaco", "T√©l√©travail"]), ignore_index=True)
```

```python
#Creation of a function to clean up the name of the location of each job offer and application to each of the datasets

import re

def trouver_ville(texte):
    for ville in fr_list:
        if re.search(r'\b{}\b'.format(ville), texte, re.IGNORECASE):
            return ville
    return "Autre"
```

```python
dfW["location"] = dfW["job_location"].apply(trouver_ville)
dfW["location"].value_counts()
```

```python
#Creation of a "job_class" column to categorize job titles

job_mapping = {
    'data scientist': ['scien'],
    'data engineer': ['engin', 'ing√©nieur'],
    'consultant data': ['consultant'],
    'master data': ['master'],
    'product owner': ['product'],
    'd√©veloppeur': ['d√©veloppeur', 'developpeur', 'ops'],
    'business analyst': ['business'],
    'data analyst': ['analys'],
    'responsable data': ['responsable', 'data'],
    'technicien data': ['technicien', 'datacenter', 'ing√©nierie', 'donn√©e'],
    'comptable': ['comptable'],
    'responsable exploitation': ['responsable', 'exploitation'],
    'responsable environnement': ['environnement'],
    'charg√© de projet': ['projet', 'chef', 'mission', 'charg√© de campagne'],
    'data manager': ['manager'],
    'big data': ['big data'],
    'charg√© de communication': ['communication'],
    'charg√© de marketing': ['marketing', 'webmarketing'],
    'responsable commercial': ['commercial', 'commerce'],
    'expert CRM': ['expert', 'CRM', 'responsable', 'crm'],
    'expert SAP': ['expert', 'SAP'],
    'data designer': ['designer'],
    'statisticien': ['statisticien', 'statistiques'],
    'expert bi': ['expert', 'bi'],
    'administrateur bases de donn√©es': ['administrateur', 'donn√©es'],
    'tech lead': ['tech', 'lead'],
    'charg√© de reporting': ['reporting'],
    'auditeur': ['auditeur'],
    'expert cloud et r√©seaux': ['cloud', 'r√©seau', 'administrateur', 'r√©seaux'],
    'expert cybers√©curit√©': ['s√©curit√©'],
    'data architect': ['architect'],
    'data steward': ['steward'],
    'autres': []
}

def map_job_class(job_title):
    job_title_lower = job_title.lower()
    for category, keywords in job_mapping.items():
        for keyword in keywords:
            if keyword in job_title_lower:
                return category
    return 'autres'

dfGlobal['job_class'] = dfGlobal['job_title'].apply(map_job_class)
```
